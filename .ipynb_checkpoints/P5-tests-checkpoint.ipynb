```python
from skimage.feature import hog
from sklearn.externals import joblib
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from scipy.ndimage.measurements import label
import time
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import cv2
import os
import glob

%matplotlib inline
```

```python
YSTART = 400
YSTOP = 656
SCALE = 1
COLOR_SPACE = 'HLS'
SPATIAL_SIZE = (16, 16)
ORIENT = 9
PIX_PER_CELL = 8
CELL_PER_BLOCK = 2
HIST_BINS = 32
HIST_RANGE = (0, 256)
HOG_CHANNEL = 'ALL'
CELLS_PER_STEP = 2
WINDOW = 64
```

```python
def save_model(svc, file='./model.pkl'):
    """
        Saves the model.
    """
    joblib.dump(svc, file)
    print("Model saved.")
    
def load_model(file='./model.pkl'):
    """
        Loads the model.
    """
    print("Loading model...")
    svc = joblib.load(file)
    print("Model loaded.")
    return svc
    
def save_scaler(X_scaler, file='./xscaler.pkl'):
    """
        Saves X scaler to disk.
    """
    joblib.dump(X_scaler, file)
    print("X scaler saved.")

def load_scaler(file='./xscaler.pkl'):
    """
        Load X scaler from disk.
    """
    print("Loading X scaler...")
    X_scaler = joblib.load(file)
    print("X scaler loaded.")
    return X_scaler

def save_features(cars_feats, notcars_feats, file='./features.pkl'):
    """
        Saves features to disk.
    """
    feats = {"cars": cars_feats, "notcars": notcars_feats}
    joblib.dump(feats, file)
    print("Features saved.")

def load_features(file='./features.pkl'):
    """
        Load features from disk.
    """
    print("Loading features...")
    feats = joblib.load(file)
    print("Features loaded.")
    return feats["cars"], feats["notcars"]
```

```python
def train_svc(X_train, y_train, X_test, y_test, name='./model.pkl'):
    """
        Trains a SVC classifier
    """
    print("Using: {} orientation, {} pixels per cell and {} cells per block".format(ORIENT, PIX_PER_CELL, CELL_PER_BLOCK))
    print("Feature vector length: {}".format(len(X_train[0])))
    init = time.time()
    if os.path.isfile(name):
        svc = load_model(name)
    else:
        svc = SVC(C=1.0, probability=True)
        svc.fit(X_train, y_train)
        save_model(svc, name)
    end = time.time()
    print("SVC trained/loaded in {} seconds".format(round(end - init, 2)))
    print("Accuracy of SVC: {}".format(svc.score(X_test, y_test)))
    return svc
```

```python
def get_scaler(cars_feats, notcars_feats, name='./xscaler.pkl'):
    """
        Fits a StandardScaler.
    """
    X = np.vstack((cars_feats, notcars_feats)).astype(np.float64)
    print('X features shape', X.shape)
    if os.path.isfile(name):
        X_scaler = load_scaler(name)
    else:
        X_scaler = StandardScaler().fit(X)
        save_scaler(X_scaler, name)
    scaled_X = X_scaler.transform(X)
    print("Scaled features: {}".format(len(scaled_X)))
    return X_scaler, scaled_X
```

```python
def split_sets(cars_feats, notcars_feats, scaled_X):
    """
        Splits data into training and tests sets:
        X_train, X_test, y_train, y_test
    """
    y = np.hstack((np.ones(len(cars_feats)), np.zeros(len(notcars_feats))))
    rand_state = np.random.randint(0, 100)
    return train_test_split(scaled_X, y, test_size=0.2, random_state=rand_state)
```

```python
def read_image(path):
    """
        Reads an image from file.
    """
    return mpimg.imread(path)
#     return cv2.imread(path)
```

```python
def load_images():
    """
        Loads the cars and not-cars images.
    """
    cars = glob.glob("./vehicles/*/*.png")
    notcars = glob.glob("./non-vehicles/*/*.png")
    print("Car samples: {}.".format(len(cars)))
    print("Not car samples: {}.".format(len(notcars)))
    return cars, notcars
```

```python
def get_hog_features(img, orient=ORIENT, pix_per_cell=PIX_PER_CELL, cell_per_block=CELL_PER_BLOCK, \
    vis=False, feature_vec=True, separate_feats=False):
    """
        Returns HOG features and visualization.
    """
    if len(img.shape) > 2 and img.shape[2] == 3:
        ch1 = img[:, :, 0]
        ch2 = img[:, :, 1]
        ch3 = img[:, :, 2]
        features = []
        if vis:
            ch1_feats, ch1_img = hog(ch1, orientations=orient, \
                pixels_per_cell=(pix_per_cell, pix_per_cell), \
                cells_per_block=(cell_per_block, cell_per_block), \
                transform_sqrt=True, visualise=vis, feature_vector=feature_vec)
            ch2_feats, ch2_img = hog(ch2, orientations=orient, \
                pixels_per_cell=(pix_per_cell, pix_per_cell), \
                cells_per_block=(cell_per_block, cell_per_block), \
                transform_sqrt=True, visualise=vis, feature_vector=feature_vec)
            ch3_feats, ch3_img = hog(ch3, orientations=orient, \
                pixels_per_cell=(pix_per_cell, pix_per_cell), \
                cells_per_block=(cell_per_block, cell_per_block), \
                transform_sqrt=True, visualise=vis, feature_vector=feature_vec)
            f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(24, 9))
            f.tight_layout()
            ax1.imshow(ch1_img, cmap='gray')
            ax2.imshow(ch2_img, cmap='gray')
            ax3.imshow(ch3_img, cmap='gray')
            plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
            plt.show()
    #         plt.close(f)
        else:
            ch1_feats = hog(ch1, orientations=orient, \
                pixels_per_cell=(pix_per_cell, pix_per_cell), \
                cells_per_block=(cell_per_block, cell_per_block), \
                transform_sqrt=True, visualise=vis, feature_vector=feature_vec)
            ch2_feats = hog(ch2, orientations=orient, \
                pixels_per_cell=(pix_per_cell, pix_per_cell), \
                cells_per_block=(cell_per_block, cell_per_block), \
                transform_sqrt=True, visualise=vis, feature_vector=feature_vec)
            ch3_feats = hog(ch3, orientations=orient, \
                pixels_per_cell=(pix_per_cell, pix_per_cell), \
                cells_per_block=(cell_per_block, cell_per_block), \
                transform_sqrt=True, visualise=vis, feature_vector=feature_vec)
        if feature_vec:
            features = np.concatenate([ch1_feats, ch2_feats, ch3_feats])
        else:
            features = np.array([ch1_feats, ch2_feats, ch3_feats])
        if separate_feats:
            return ch1_feats, ch2_feats, ch3_feats
        else:
            return np.array(features)
    else:
        if vis:
            hog_feats, hog_img = hog(img, orientations=orient, \
                pixels_per_cell=(pix_per_cell, pix_per_cell), \
                cells_per_block=(cell_per_block, cell_per_block), \
                transform_sqrt=True, visualise=vis, feature_vector=feature_vec)
            plt.imshow(hog_img, cmap='gray')
            plt.show()
        else:
            hog_feats = hog(img, orientations=orient, \
                pixels_per_cell=(pix_per_cell, pix_per_cell), \
                cells_per_block=(cell_per_block, cell_per_block), \
                transform_sqrt=True, visualise=vis, feature_vector=feature_vec)
        return hog_feats
```

```python
def bin_spatial(img, size=SPATIAL_SIZE, vis=False):
    """
        Returns a feature vector.
    """
    features = cv2.resize(img, size).ravel()
    if vis:
        plt.plot(features)
        plt.show()
    return features
```

```python
def color_hist(img, nbins=HIST_BINS, bins_range=HIST_RANGE, channel='ALL', vis=False):
    """
        Computes the histogram of the color channels separately.
    """
    if channel == 'ALL':
        channel1_hist = np.histogram(img[:, :, 0], bins=nbins, range=bins_range)
        channel2_hist = np.histogram(img[:, :, 1], bins=nbins, range=bins_range)
        channel3_hist = np.histogram(img[:, :, 2], bins=nbins, range=bins_range)
        features = np.concatenate((channel1_hist[0], channel2_hist[0], channel3_hist[0]))
    else:
        features = np.histogram(img[:, :, channel], bins=nbins, range=bins_range)
    if vis:
        plt.plot(features)
        plt.show()
    return features
```

```python
def to_color_space(img, color_space, separate_channels=False, vis=False):
    if color_space != 'RGB':
        if color_space == 'HSV':
            feature_image = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
        elif color_space == 'LUV':
            feature_image = cv2.cvtColor(img, cv2.COLOR_RGB2LUV)
        elif color_space == 'HLS':
            feature_image = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
        elif color_space == 'YUV':
            feature_image = cv2.cvtColor(img, cv2.COLOR_RGB2YUV)
        elif color_space == 'YCrCb':
            feature_image = cv2.cvtColor(img, cv2.COLOR_RGB2YCrCb)
        elif color_space == 'Gray':
            feature_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    else: feature_image = np.copy(img)
    if vis:
        plt.imshow(feature_image)
        plt.show()
    if separate_channels:
        return feature_image, feature_image[:, :, 0], feature_image[:, :, 1], feature_image[:, :, 2]
    else:
        return feature_image
```

```python
def compose_img(image, vis=False):
    img = np.copy(image)
    hls = to_color_space(img, 'HLS', vis=vis)
    ycc = to_color_space(img, 'YCrCb', vis=vis)
    gray = to_color_space(img, 'Gray', vis=vis)
    ch_l = hls[:,:,1]
    ch_y = ycc[:,:,0]
    img[:,:,0] = ch_l
    img[:,:,1] = ch_y
    img[:,:,2] = gray
    return img, ch_l, ch_y, gray
```

```python
def extract_features(img, color_space=COLOR_SPACE, orient=ORIENT, pix_per_cell=PIX_PER_CELL, \
    cell_per_block=CELL_PER_BLOCK, hog_channel=HOG_CHANNEL, spatial_size=SPATIAL_SIZE, hist_bins=HIST_BINS, \
    hist_range=HIST_RANGE, vis=False):
    """
        Converts color space and extracts hog features from a single image window.
    """
    feature_image = to_color_space(img, color_space)
    if vis:
        plt.imshow(feature_image)
        plt.show()
    features = []
    hog_features = get_hog_features(feature_image, orient, pix_per_cell, cell_per_block, vis, True)
    features.append(hog_features)
    spatial_features = bin_spatial(feature_image, spatial_size, vis=vis)
    features.append(spatial_features)
    hist_features = color_hist(feature_image, hist_bins, hist_range, vis=vis)
    features.append(hist_features)
    return np.concatenate(features)
```

```python
def list_extract_features(images, vis=False):
    features = []
    for file in images:
        image = read_image(file)
        img_features = extract_features(image, vis=vis)
        features.append(img_features)
    return features
```

```python
def slide_window(image, svc, X_scaler, ystart=YSTART, ystop=YSTOP, cells_per_step=2, window_scale=SCALE, \
    window=WINDOW, color_space=COLOR_SPACE, vis=False):
    """
        Returns a list of sliding windows.
    """
    img = np.copy(image)
    img = img.astype(np.float32) / 255
    img_tosearch = img[ystart:ystop, :, :]
    if window_scale != 1:
        imshape = img_tosearch.shape
        img_tosearch = cv2.resize(img_tosearch, (np.int(imshape[1] / window_scale), np.int(imshape[0] / window_scale)))
        if vis:
            plt.imshow(img_tosearch)
            plt.show()
#     ctrans_tosearch, ch1, ch2, ch3 = to_color_space(img_tosearch, color_space, separate_channels=True, vis=vis)
    ctrans_tosearch, ch1, ch2, ch3 = compose_img(img_tosearch, vis=vis)
    if vis:
        f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(24, 9))
        f.tight_layout()
        ax1.imshow(ch1)
        ax2.imshow(ch2)
        ax3.imshow(ch3)
        plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
        plt.show()
#         plt.close(f)
    hog1, hog2, hog3 = get_hog_features(ctrans_tosearch, vis=vis, feature_vec=False, separate_feats=True)
    nxblocks = (ch1.shape[1] // PIX_PER_CELL) - CELL_PER_BLOCK + 1
    nyblocks = (ch1.shape[0] // PIX_PER_CELL) - CELL_PER_BLOCK + 1
    nblocks_per_window = (window // PIX_PER_CELL) - CELL_PER_BLOCK + 1
    nxsteps = (nxblocks - nblocks_per_window) // cells_per_step
    nysteps = (nyblocks - nblocks_per_window) // cells_per_step
    bboxes = []
    for xb in range(nxsteps):
        for yb in range(nysteps):
            ypos = yb*cells_per_step
            xpos = xb*cells_per_step
            hog_feat1 = hog1[ypos:ypos+nblocks_per_window, xpos:xpos+nblocks_per_window].ravel()
            hog_feat2 = hog2[ypos:ypos+nblocks_per_window, xpos:xpos+nblocks_per_window].ravel()
            hog_feat3 = hog3[ypos:ypos+nblocks_per_window, xpos:xpos+nblocks_per_window].ravel()
            hog_features = np.hstack((hog_feat1, hog_feat2, hog_feat3))
            xleft = xpos  * PIX_PER_CELL
            ytop = ypos * PIX_PER_CELL
            subimg = cv2.resize(ctrans_tosearch[ytop:ytop+window, xleft:xleft+window], (64, 64))
            spatial_features = bin_spatial(subimg, vis=False)
            hist_features = color_hist(subimg, vis=False)
            all_features = np.hstack((spatial_features, hist_features, hog_features)).reshape(1, -1)
            test_features = X_scaler.transform(all_features)
            test_prediction = svc.predict(test_features)
            if test_prediction == 1:
                xbox_left = np.int(xleft)
                ytop_draw = np.int(ytop)
                win_draw = np.int(window)
                bboxes.append(((xbox_left, ytop_draw + ystart), (xbox_left + win_draw, ytop_draw + win_draw + ystart)))
    return bboxes
```

```python
def draw_boxes(img, bboxes, color=(255, 0, 0), thick=6):
    """
        Draws boxes on the image.
    """
    imcopy = np.copy(img)
    for bbox in bboxes:
        cv2.rectangle(imcopy, bbox[0], bbox[1], color, thick)
    return imcopy
```

```python
def add_heat(heatmap, bbox_list):
    """
        Adds "heat" to a list of bounding boxes.
    """
    for box in bbox_list:
        heatmap[box[0][1]:box[1][1], box[0][0]:box[1][0]] += 1
    return heatmap
```

```python
def apply_threshold(heatmap, threshold):
    """
        Applies a threshold on a heatmap.
    """
    heatmap[heatmap <= threshold] = 0
    return heatmap
```

```python
def draw_labeled_bboxes(img, labels):
    """
        Draws labeled boxes over detected cars.
    """
    for car_number in range(1, labels[1] + 1):
        nonzero = (labels[0] == car_number).nonzero()
        nonzeroy = np.array(nonzero[0])
        nonzerox = np.array(nonzero[1])
        bbox = ((np.min(nonzerox), np.min(nonzeroy)), (np.max(nonzerox), np.max(nonzeroy)))
        cv2.rectangle(img, bbox[0], bbox[1], (0, 0, 255), 6)
    return img
```

```python
def train_pipeline(color_space=COLOR_SPACE):
    """
        Trains the SVM
    """
    model_name = './' + color_space + '_model.pkl'
    scaler_name = './' + color_space + '_scaler.pkl'
    features_name = './' + color_space + '_features.pkl'
    cars, notcars = load_images()
    if (os.path.isfile(features_name)):
        cars_feats, notcars_feats = load_features()
    else:
        cars_feats = list_extract_features(cars)
        notcars_feats = list_extract_features(notcars)
        save_features(cars_feats, notcars_feats, features_name)
    X_scaler, scaled_X = get_scaler(cars_feats, notcars_feats, scaler_name)
    X_train, X_test, y_train, y_test = split_sets(cars_feats, notcars_feats, scaled_X)
    svc = train_svc(X_train, y_train, X_test, y_test, name=model_name)
    return svc, X_scaler
```

```python
def pipeline():
    model_name = './' + COLOR_SPACE + '_model.pkl'
    scaler_name = './' + COLOR_SPACE + '_scaler.pkl'
    image = read_image(os.path.join('test_images', 'test1.jpg'))
    svc, X_scaler = train_pipeline()
    # svc = load_model(model_name)
    # X_scaler = load_scaler(scaler_name)
    bboxes = slide_window(image, svc, X_scaler, vis=True)
    bboxes += slide_window(image, svc, X_scaler, window_scale=1.5, vis=True)
    bboxes += slide_window(image, svc, X_scaler, ystart=400, ystop=500, window_scale=.75, vis=True)
    boxes_img = draw_boxes(image, bboxes)
    plt.imshow(boxes_img)
    plt.show()
    heat = np.zeros_like(image[:, :, 0]).astype(np.float)
    heat = add_heat(heat, bboxes)
    heat = apply_threshold(heat, 1)
    heatmap = np.clip(heat, 0, 255)
    plt.imshow(heatmap, cmap='hot')
    plt.show()
    labels = label(heatmap)
    result = draw_labeled_bboxes(image, labels)
    plt.imshow(result)
    plt.show()
    print('End')
```
