```python
from skimage.feature import hog
from sklearn.externals import joblib
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from scipy.ndimage.measurements import label
import numpy as np
import cv2
import os

%matplotlib inline
```

```python
YSTART = 400
YSTOP = 656
SCALE = 1
COLOR_SPACE = 'RGB'
SPATIAL_SIZE = (32, 32)
ORIENT = 9
PIX_PER_CELL = 8
CELL_PER_BLOCK = 2
HIST_BINS = 32
HIST_RANGE = (0, 256)
HOG_CHANNEL = 'ALL'
CELLS_PER_STEP = 2
WINDOW = 64
```

```python
def read_image(path):
    """
        Reads an image from file.
    """
    return mpimg.imread(path)
```

```python
def get_hog_features(img, orient=ORIENT, pix_per_cell=PIX_PER_CELL, cell_per_block=CELL_PER_BLOCK, vis=False, feature_vec=True):
    """
        Returns HOG features and visualization.
    """
    if vis:
        features, hog_image = hog(img, orientations=orient, \
            pixels_per_cell=(pix_per_cell, pix_per_cell), \
            cells_per_block=(cell_per_block, cell_per_block), \
            transform_sqrt=True, visualise=vis, feature_vector=feature_vec)
        return features, hog_image
    else:
        features = hog(img, orientations=orient, \
            pixels_per_cell=(pix_per_cell, pix_per_cell), \
            cells_per_block=(cell_per_block, cell_per_block), \
            transform_sqrt=True, visualise=vis, feature_vector=feature_vec)
        return features
```

```python
def bin_spatial(img, size=SPATIAL_SIZE):
    """
        Returns a feature vector.
    """
    features = cv2.resize(img, size).ravel()
    return features
```

```python
def color_hist(img, nbins=HIST_BINS, bins_range=HIST_RANGE, channel='ALL'):
    """
        Computes the histogram of the color channels separately.
    """
    if channel == 'ALL':
        channel1_hist = np.histogram(img[:, :, 0], bins=nbins, range=bins_range)
        channel2_hist = np.histogram(img[:, :, 1], bins=nbins, range=bins_range)
        channel3_hist = np.histogram(img[:, :, 2], bins=nbins, range=bins_range)
        hist_features = np.concatenate((channel1_hist[0], channel2_hist[0], channel3_hist[0]))
    else:
        hist_features = np.histogram(img[:, :, channel], bins=nbins, range=bins_range)
    return hist_features
```

```python
def to_color_space(img, color_space):
    if color_space != 'RGB':
        if color_space == 'HSV':
            feature_image = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
        elif color_space == 'LUV':
            feature_image = cv2.cvtColor(img, cv2.COLOR_RGB2LUV)
        elif color_space == 'HLS':
            feature_image = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
        elif color_space == 'YUV':
            feature_image = cv2.cvtColor(img, cv2.COLOR_RGB2YUV)
        elif color_space == 'YCrCb':
            feature_image = cv2.cvtColor(img, cv2.COLOR_RGB2YCrCb)
    else: feature_image = np.copy(img)
    yuv = cv2.cvtColor(img, cv2.COLOR_RGB2YUV)
    ycc = cv2.cvtColor(img, cv2.COLOR_RGB2YCrCb)
    hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    feature_image[:, :, 1] = yuv[:, :, 0]
    feature_image[:, :, 2] = hsv[:, :, 2]
    return feature_image
```

```python
def extract_features(img, color_space=COLOR_SPACE, orient=ORIENT, pix_per_cell=PIX_PER_CELL, \
    cell_per_block=CELL_PER_BLOCK, hog_channel=HOG_CHANNEL):
    """
        Converts color space and extracts hog features from a single image window.
    """
    if color_space != 'RGB':
        if color_space == 'HSV':
            feature_image = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
        elif color_space == 'LUV':
            feature_image = cv2.cvtColor(img, cv2.COLOR_RGB2LUV)
        elif color_space == 'HLS':
            feature_image = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
        elif color_space == 'YUV':
            feature_image = cv2.cvtColor(img, cv2.COLOR_RGB2YUV)
        elif color_space == 'YCrCb':
            feature_image = cv2.cvtColor(img, cv2.COLOR_RGB2YCrCb)
    else: feature_image = np.copy(img)
    plt.imshow(feature_image)
    plt.show()
    if hog_channel == 'ALL':
        hog_features = []
        for channel in range(feature_image.shape[2]):
            hog_f, hog_img = get_hog_features(feature_image[:, :, channel], \
                    orient, pix_per_cell, cell_per_block, True, True)
            plt.imshow(hog_img, cmap='gray')
            plt.show()
            hog_features.append(hog_f)
    else:
        hog_features, hog_img = get_hog_features(feature_image[:, :, \
            hog_channel], orient, pix_per_cell, cell_per_block, True, True)
        plt.imshow(hog_img, cmap='gray')
        plt.show()
    return feature_image, hog_features
```

```python
def slide_window(image, svc, X_scaler, ystart=YSTART, ystop=YSTOP, color_space=COLOR_SPACE, orient=ORIENT, \
    pix_per_cell=PIX_PER_CELL, cell_per_block=CELL_PER_BLOCK, hog_channel=HOG_CHANNEL, \
    cells_per_step=CELLS_PER_STEP, window=WINDOW):
    """
        Returns a list of sliding windows.
    """
    img = image.astype(np.float32) / 255
    img_tosearch = img[ystart:ystop, :, :]
    ctrans_tosearch = to_color_space(img_tosearch, color_space)
    ch1 = ctrans_tosearch[:, :, 0]
    ch2 = ctrans_tosearch[:, :, 1]
    ch3 = ctrans_tosearch[:, :, 2]
    plt.imshow(ch1)
    plt.show()
    plt.imshow(ch2)
    plt.show()
    plt.imshow(ch3)
    plt.show()
    hog1, hog1_img = get_hog_features(ch1, vis=True, feature_vec=False)
    hog2, hog2_img = get_hog_features(ch2, vis=True, feature_vec=False)
    hog3, hog3_img = get_hog_features(ch3, vis=True, feature_vec=False)
    plt.imshow(hog1_img)
    plt.show()
    plt.imshow(hog2_img)
    plt.show()
    plt.imshow(hog3_img)
    plt.show()
    nxblocks = (ch1.shape[1] // pix_per_cell) - cell_per_block + 1
    nyblocks = (ch1.shape[0] // pix_per_cell) - cell_per_block + 1
    nblocks_per_window = (window // pix_per_cell) - cell_per_block + 1
    nxsteps = (nxblocks - nblocks_per_window) // cells_per_step
    nysteps = (nyblocks - nblocks_per_window) // cells_per_step
    bboxes = []
    for xb in range(nxsteps):
        for yb in range(nysteps):
            ypos = yb*cells_per_step
            xpos = xb*cells_per_step
            hog_feat1 = hog1[ypos:ypos+nblocks_per_window, xpos:xpos+nblocks_per_window].ravel()
            hog_feat2 = hog2[ypos:ypos+nblocks_per_window, xpos:xpos+nblocks_per_window].ravel()
            hog_feat3 = hog3[ypos:ypos+nblocks_per_window, xpos:xpos+nblocks_per_window].ravel()
            hog_features = np.hstack((hog_feat1, hog_feat2, hog_feat3))
            xleft = xpos*pix_per_cell
            ytop = ypos*pix_per_cell
            subimg = cv2.resize(ctrans_tosearch[ytop:ytop+window, xleft:xleft+window], (64, 64))
            spatial_features = bin_spatial(subimg)
#             plt.plot(spatial_features)
#             plt.show()
            hist_features = color_hist(subimg)
#             plt.plot(hist_features)
#             plt.show()
            all_features = np.hstack((spatial_features, hist_features, hog_features)).reshape(1, -1)
            test_features = X_scaler.transform(all_features)
            test_prediction = svc.predict(test_features)
            if test_prediction == 1:
                xbox_left = np.int(xleft)
                ytop_draw = np.int(ytop)
                win_draw = np.int(window)
                bboxes.append(((xbox_left, ytop_draw + ystart), (xbox_left + win_draw, ytop_draw + win_draw + ystart)))
    return bboxes
```

```python
def draw_boxes(img, bboxes, color=(0, 0, 255), thick=6):
    """
        Draws boxes on the image.
    """
    imcopy = np.copy(img)
    for bbox in bboxes:
        cv2.rectangle(imcopy, bbox[0], bbox[1], color, thick)
    return imcopy
```

```python
def add_heat(heatmap, bbox_list):
    """
        Adds "heat" to a list of bounding boxes.
    """
    for box in bbox_list:
        heatmap[box[0][1]:box[1][1], box[0][0]:box[1][0]] += 1
    return heatmap
```

```python
def apply_threshold(heatmap, threshold):
    """
        Applies a threshold on a heatmap.
    """
    heatmap[heatmap <= threshold] = 0
    return heatmap
```

```python
def draw_labeled_bboxes(img, labels):
    """
        Draws labeled boxes over detected cars.
    """
    for car_number in range(1, labels[1] + 1):
        nonzero = (labels[0] == car_number).nonzero()
        nonzeroy = np.array(nonzero[0])
        nonzerox = np.array(nonzero[1])
        bbox = ((np.min(nonzerox), np.min(nonzeroy)), (np.max(nonzerox), np.max(nonzeroy)))
        cv2.rectangle(img, bbox[0], bbox[1], (0, 0, 255), 6)
    return img
```

```python
def load_model(file='./model.pkl'):
    """
        Loads the model.
    """
    print("Loading model...")
    svc = joblib.load(file)
    print("Model loaded.")
    return svc

def load_scaler(file='./xscaler.pkl'):
    """
        Load X scaler from disk.
    """
    print("Loading X scaler...")
    X_scaler = joblib.load(file)
    print("X scaler loaded.")
    return X_scaler
```

```python
image = read_image(os.path.join('test_images', 'test1.jpg'))
X_scaler = load_scaler()
svc = load_model()
bboxes = slide_window(image, svc, X_scaler)
boxes_img = draw_boxes(image, bboxes)
plt.imshow(boxes_img)
plt.show()
heat = np.zeros_like(image[:, :, 0]).astype(np.float)
heat = add_heat(heat, bboxes)
heat = apply_threshold(heat, 1)
heatmap = np.clip(heat, 0, 255)
plt.imshow(heatmap, cmap='hot')
plt.show()
labels = label(heatmap)
result = draw_labeled_bboxes(image, labels)
plt.imshow(result)
plt.show()
```
